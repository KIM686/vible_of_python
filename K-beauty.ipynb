```python
df["연도"] =df["기간"].map(lambda x: int(x.split()[0]))
df.head()
```

```python
"2019 4/4 p".split()[1].split("/") #공백과 /을 기준으로 데이터를 split하여 4/4에 있는 4와 4를 분리시킴
```

```python
"2019 4/4)".split()[1].split("/")[0] #4/4에 있는 앞의 4를 가져옵니다.
```

```python
df["분기"] =df["기간"].map( lambda x: int( x.split()[1].split("/")[0]              )      )
df.head()
```

```python
#2016년이전데이터 결측치가 많고 -으로 기록, -으로 표시된 값은 NaN으로 대치
df["백만원"].replace("-",np.nan)
```

```python
#numpy 로드하지않았으면 pandas안에 있는 mp.nan 쓸수있다
df["백만원"] =df["백만원"].replace("-",np.nan).astype(float)
df["백만원"]
```

```python
df.info()
```

```python
df = df[ (df["국가(대륙)별"]!= "합계") & (df["상품군별"] != "합계" ) ].copy()
```
